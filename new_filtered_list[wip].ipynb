```python
import pandas as pd
```

```python
nura = pd.read_csv('/content/nura_following_df.csv')
luci = pd.read_csv('/content/miss_luci_following_df.csv')
natalie = pd.read_csv('/content/princess_natalie_following_df.csv')
last = pd.read_csv('/content/Twitter Database of Creators [Stephh] - awesome_twitter_df.csv')

```

```python
#merging all of the creator following dataframes :)
nura_luci = pd.merge(luci, nura, on='screenName', how='outer')
all = pd.merge(nura_luci, natalie, on='screenName', how='outer')

#now clean duplicates from list...
duplicates = all.duplicated(subset=['screenName'])

new_filtered_list = all[~all['profileURL'].isin(last['profileURL'])]
```
