```python
import pandas as pd
```

```python
nura = pd.read_csv('/content/nura_following_df.csv')
luci = pd.read_csv('/content/miss_luci_following_df.csv')
natalie = pd.read_csv('/content/princess_natalie_following_df.csv')
last = pd.read_csv('/content/Twitter Database of Creators [Stephh] - awesome_twitter_df.csv')

```

```python
#combining all of the creator following dataframes :)
combined = pd.concat([nura, natalie, luci], ignore_index=True)
combined = combined.drop_duplicates(subset="profileUrl")
combined.shape, combined.head()
```

```python
last.head()
```

```python
# Clean URLs by removing 'http://', 'https://'
combined['profileUrl'] = combined['profileUrl'].str.replace('http://', '').str.replace('https://', '')
last['Profile URL'] = last['Profile URL'].str.replace('http://', '').str.replace('https://', '')

# Check for missing values in 'profileUrl' and replace with an empty string
last['Profile URL'].fillna('', inplace=True)

# Initialize an empty list to store cleaned URLs for 'last'
clean_last = []

# Loop through each URL in 'last' and clean it
for url in last['Profile URL']:
    cleaned_url = url.split('//')[-1].split('?')[0].split('#')[0]
    clean_last.append(cleaned_url)

last['cleaned_profile'] = last['Profile URL'].apply(
    lambda url: url.split('//')[-1].split('?')[0].split('#')[0]
)

# Strip and convert to lowercase for better comparison
combined['profileUrl'] = combined['profileUrl'].str.strip().str.lower()
last['cleaned_profile'] = last['cleaned_profile'].str.strip().str.lower()

# Filter rows in 'combined' that are also in 'last'
common_last = combined[combined['profileUrl'].isin(last['cleaned_profile'])]

# Update 'combined' to exclude rows that are in 'last'
combined = combined[~combined['profileUrl'].isin(last['cleaned_profile'])]
combined.to_csv('new_list.csv', index=False)
from google.colab import files
files.download('new_list.csv')

```
